```python
#akyork
#written to analyze the 16 HMO glycans provided by Ben

import sys
import __init__
import json_utility
from glypy.io import glycoct, iupac
import extract_motif 
import customize_motif_vec 
import glycan_io
import pandas as pd
import numpy as np
```

```python
#run motif search on 16 HMO glycans

#loads the json into dictionary form
hmo_dict = json_utility.load_json("/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/glycanHMO_dict.json")

#turns hmo dict to glycoCT dict
hmoCT_dict = {}
for items in hmo_dict.items():
    hmoCT_dict[items[0]] = glycoct.loads(items[1])

#gets motifs and saves to json file
hmoMotif_dict = extract_motif.get_motif_pip(hmoCT_dict, gly_len=22, output_file="/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/glycanHMO_motif_dict.json")

#turns to non-string form?                                    
hmoMotif_dict = glycan_io.glycan_str_to_glycan(json_utility.load_json("/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/glycanHMO_motif_dict.json"))                     
```

```python

#merges the motif dict with the glycan dict because full chains are still considered motifs
master_motif_dict = customize_motif_vec.merge_motif_dict_pipe(hmoMotif_dict, __init__.merged_motif_dict_addr)

#produces the list of motifs as they are represented in the table
# i.e. 0 = motif 1, 1 = motif2, etc...
motif_vector = glycan_io.motif_dict_to_motif_vec(master_motif_dict)

#creats a file for motif list
file = open('/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/Data/motif_list.txt', 'w')
for item in motif_vector:
  file.write("%s\n" % item)


#print(motif_vector[0])
```

```python

#vectorizes the occurence of the motifs per glycan
motif_occurance_vector_dict = customize_motif_vec.motif_matching_wrapper(master_motif_dict, hmoMotif_dict, __init__.output_matched_dict_addr)
```

```python
#shows occurance of motifs per glycans
motif_data = pd.DataFrame(data=motif_occurance_vector_dict)
```

```python
#translation table for changing the ID to shorthand name

translation_table = {
    
    'G45827GY':'LNT',
    'G48059CD':'LNnT',
    'G10422IZ':'''2'FL''',
    'G06210XB':'3FL',
    'G91237TK':'''3'SL''',
    'G01650PH':'LNFP I',
    'G98173LG':'LNFP II',
    'G83916HL':'LNFP III',
    'G19017MP':'LSTb',
    'G72506RN':'LSTc',
    'G38710SX':'DSLNT',
    'G70115XG':'DFLNT',
    'G24504JY':'FLNH',
    'G63053GR':'DFLNH',
    'G47928KI':'DSLNH',
    'FDSLNH':'FDSLNH'
    
}

#gets the glycan abundance data set
glycan_data = pd.read_csv('/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/Data/Data12.csv')


#changes the column names of motif data to glycan data nomenclature
translation_list = []
for i in list(motif_data):
    if translation_table[i] in list(glycan_data):
        translation_list.append(translation_table[i])
motif_data.columns = translation_list 

```

```python
#gets each data set in similar sorted order
sorted_motif_data = motif_data.sort_index(axis=1)

################################################################
#specific to the data set itself, only uses the 16 glycan columns
################################################################
trimed_glycan_data = glycan_data.iloc[:,3:19]
trimed_sorted_glycan_data = trimed_glycan_data.sort_index(axis=1)

```

```python
#matrix multiplication on the two sorted/trimed data sets to produce the motif abundance

motif_abundance = pd.DataFrame(data=np.dot(sorted_motif_data, trimed_sorted_glycan_data.transpose()).transpose())

################################################################
#             sets column names to motifs
################################################################
#motif_abundance.columns = motif_vector

```

```python
#replace column one with the first three columns from original Data12 to match Data12 format

info_columns = glycan_data.iloc[:,0:3]

result = pd.concat([info_columns, motif_abundance], axis=1, sort=False)

result.to_csv('/Users/AustinYork/Documents/LewisLab/GlycanMotif/GlyCompare-master/Data/motif_abundance.csv')
```
